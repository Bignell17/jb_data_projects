UDF

```python
def chunk_text(text, size=300):
    return [
        text[i:i+size]
        for i in range(0, len(text), size)
    ]
```

```python
chunks = chunk_text(
    "Databricks is a lakehouse platform for data and AI.")
print(chunks)
```
