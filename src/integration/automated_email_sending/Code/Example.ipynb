## Transformation

```python
spark = SparkSession.builder.getOrCreate()

# Employees DataFrame
employees_data = [
    (1, 'Alice', 55000, 'HR'),
    (2, 'Bob', 62000, 'Finance'),
    (3, 'Charlie', 58000, 'IT'),
    (4, 'Diana', 70000, 'Marketing'),
    (5, 'Ethan', 54000, 'IT')
]

employees_columns = ["emp_id", "name", "salary", "department"]

df_employees = spark.createDataFrame(employees_data, employees_columns)

# Customers DataFrame
customers_data = [
    (1, 'Jerry', 1001, 'Water'),
    (2, 'Alison', 1002, 'Butter'),
    (3, 'Sandra', 1003, 'Eggs'),
    (4, 'Barry', 1004, 'Milk'),
    (5, 'Steve', 1005, 'Bread')
]

customers_columns = ["cust_id", "name", "order_id", "item_name"]

df_customers = spark.createDataFrame(customers_data, customers_columns)
```
